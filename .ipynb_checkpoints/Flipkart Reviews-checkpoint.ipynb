```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
base_url = 'https://www.flipkart.com/flipkart-smartbuy-1000-w-dry-iron/product-reviews/itmf3kcr58zcbh6p?pid=IRNF28ETGB7HHRWE&lid=LSTIRNF28ETGB7HHRWEYGG9WM&marketplace=FLIPKART99&page='
```

```python
j = 1
review_data = []

for i in range(1, 130):
    url = f'{base_url}{i}'
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text, 'lxml')
    reviews = soup.find_all('div', class_ = 'col EPCmJX Ma1fCG')
    product_name_tag = soup.find('a', class_= 'wjcEIp AbG6iz')
    if product_name_tag:
        product_name = product_name_tag.text.strip()

    for review in reviews:
        name_tag = review.find_all('p', class_='_2NsDsF AwS1CA')
        stars_tag = review.find('div', class_='XQDdHH Ga3i8K')
        review_tag = review.find('p', class_='z9E0IG')
        review_text_tag = review.find('div', class_='ZmyHeo')
        date_tag = review.find_all('p', class_='_2NsDsF')
        

        if name_tag and stars_tag and review_text_tag and review_tag:
            name = name_tag[0].text.strip()
            stars = stars_tag.text.strip()
            review = review_tag.text.strip()
            review_text = review_text_tag.text.strip()
            date = date_tag[1].text.strip()
            if 'READ MORE' in review_text:
                review_text = review_text[:-9]

            print(j, ":")
            print(f'Product Name: {product_name}')
            print(f'Name: {name}')
            print(f'Stars: {stars}')
            print(f'Review Title: {review}')
            print(f'Review Text: {review_text}')
            print(f'Date: {date}')
            print()
            j += 1
            review_data.append({'Product Name': product_name, 'Name': name, 'Rating': stars, 'Review Title': review, 'Review Text': review_text, 'Date': date})

print('Data collection completed.')


```

```python
df = pd.DataFrame(review_data)
df  
```

```python
df.to_csv('Iron Box Flipkart Reviews.csv', index=False)
```
