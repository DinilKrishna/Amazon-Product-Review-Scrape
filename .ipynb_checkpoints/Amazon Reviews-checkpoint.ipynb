```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
base_url = 'https://www.amazon.in/LOreal-Paris-Excellence-Natural-Darkest/product-reviews/B006QHB88O/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber='
```

```python
j = 1
review_data = []
for i in range(1, 11):
    url = f'{base_url}{i}'
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text, 'lxml')
    reviews = soup.find_all('div', class_ = 'a-section celwidget')
    for review in reviews:
        name = review.find('span', class_ = 'a-profile-name').text
        stars = review.find('span', class_ = 'a-icon-alt').text
        review = review.find('span', class_ = 'a-size-base review-text review-text-content').span.text
        print(j,":")
        print(f'Name: {name}')
        print(f'Stars: {stars}')
        print(f'Review: {review}')
        print()
        j += 1
        review_data.append({'Name': name, 'Stars': stars, 'Review': review})

    
```

```python
df = pd.DataFrame(review_data)
df    
```

```python
df.to_csv('Loreal_Paris_reviews.csv', index=False)
```
